```python
import pandas as pd
import numpy as np
import pyarrow.parquet as pq
```

```python
prog_df_tbl = pq.read_table("/Users/mthota/Dropbox/Data/Data/Program/correct File/parquetwithHeaders/part-00000-450b76c4-8d60-4858-8fce-4ec9905e0def-c000.snappy.parquet")
```

```python
prog_df = prog_df_tbl.to_pandas()

prog_df= prog_df[(prog_df.EVENT_DATE>='2018-01-01')& (prog_df.EVENT_DATE!='NULL')]
prog_df.to_csv("/Users/mthota/Dropbox/Data/Data/Program/correct File/program_2018.csv")
prog_df.replace("NULL", np.nan,inplace=True)
prog_df.info()
```

```python
print(prog_df['SUBCATEGORY_ID'].nunique())
print(prog_df["SUBCATEGORY_ID"].value_counts())
prog_df["SUBCATEGORY_ID"].value_counts().plot(kind="bar",figsize=(12,5))
print(prog_df["SUBCATEGORY_ID"].value_counts())
```

```python
prog_df["CATEGORY_ID"].unique()
prog_df["CATEGORY_ID"].value_counts()
prog_df["CATEGORY_ID"].value_counts().plot(kind="bar")

def func(row):
    if row['CATEGORY_ID'] == '1':
        return 'movies'
    elif row['CATEGORY_ID'] == '2':
        return 'sports' 
    elif row['CATEGORY_ID'] == '3':
        return 'kids' 
    elif row['CATEGORY_ID'] == '4':
        return 'news' 
    elif row['CATEGORY_ID'] == '5':
        return 'tvshows' 
    elif row['CATEGORY_ID'] == '8':
        return 'foodshows' 
    elif row['CATEGORY_ID'] == '9':
        return 'music' 

prog_df['CATEGORY'] = prog_df.apply(func, axis=1)
```

```python
prog_df.CATEGORY.value_counts()
prog_df["CATEGORY"].value_counts().plot(kind="bar")
```

```python
df_cat_sub_cat = pd.DataFrame(prog_df[['CATEGORY','SUBCATEGORY_ID','PROGRAM_ID']])

df_cat_group= pd.DataFrame(df_cat_sub_cat.groupby(['SUBCATEGORY_ID','CATEGORY']).count()['PROGRAM_ID'])
df_cat_sub_cat_1  = df_cat_sub_cat[(df_cat_sub_cat.CATEGORY=='tvshows')|(df_cat_sub_cat.CATEGORY=='kids')|(df_cat_sub_cat.CATEGORY=='foodshows')]
pd.DataFrame(df_cat_sub_cat_1.groupby(['SUBCATEGORY_ID']).count()['PROGRAM_ID']).sort_values('PROGRAM_ID',ascending=False).plot(kind="bar",figsize=(12,5))
```

```python
df_cat_sub_cat_2  = df_cat_sub_cat[df_cat_sub_cat.CATEGORY=='sports']
pd.DataFrame(df_cat_sub_cat_2.groupby(['SUBCATEGORY_ID']).count()['PROGRAM_ID']).sort_values('PROGRAM_ID',ascending=False).plot(kind="bar",figsize=(12,5))

## business quedstions skew in data for genre shld we consier all genres
```
